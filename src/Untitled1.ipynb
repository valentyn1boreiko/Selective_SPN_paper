```python
from scipy.interpolate import interp1d
from matplotlib import pyplot as plt
from scipy.integrate import trapz
import numpy as np
import pandas as pd
import os


plt.style.use('fivethirtyeight')

def plot_value(val, subdirs = ""):

    fig, ax = plt.subplots(figsize=(80,60))

    path = "../out/nltcs"

    df_arr = []
    
    if subdirs != "":
        for subdir, dirs, files in os.walk(path):
            if subdir in subdirs:
                print val
                print subdir
                for file in files[::-1]:
                    df_temp = pd.read_csv(os.path.join(subdir, file))
                    to_plot = val
                    x = "iteration"
                    if file == "configuration.csv":

                        label_ = df_temp.values

                    if file == "history.csv":

                        label_ =  str(label_)
                        for col in ["train_bd","val_bd","test_bd","train_llh","val_llh","test_llh","num_nodes"]:
                            label_ = label_ +"max_" + col + " = " + str(max(df_temp[col])) + " "
                        ax.plot(df_temp["iteration"],df_temp[to_plot], label = label_)

        ax.legend(loc='lower right')

        plt.savefig(val+'.svg')
        
        plt.close()

    else:
        
        subdirs = []
        
        for subdir, dirs, files in os.walk(path):
            for file in files[::-1]:
                df_temp = pd.read_csv(os.path.join(subdir, file))
                to_plot = val
                x = "iteration"
                if file == "configuration.csv":

                    label_ = df_temp.values

                if file == "history.csv":

                    if len(df_arr) > 0:
                        ints_arr = []
                        for df_ in df_arr:

                            l1 = df_temp[x].tolist()
                            l2 = df_[x].tolist()
                            id_from = l1.index(filter(lambda x: x > min(l2) , l1)[0])
                            id_to = l1.index(filter(lambda x: x < max(l2) , l1)[-1])
                            xx = df_temp[x][id_from:id_to]
                            int1 = interp1d(df_[x], df_[to_plot])(xx)
                            int2 = df_temp[to_plot][id_from:id_to]

                            #ints_arr.append(trapz(np.abs(int1 - int2), xx) / len(xx))
                            #ints_arr.append(max(np.abs(int1-int2))/max(abs(min(int1)),abs(min(int2))))
                            last_id = len(int1) - 1
                            ints_arr.append(np.abs(int1[last_id]-int2[last_id]))
                    else:
                        ints_arr = [0]

                    df_arr.append(df_temp)
                    print min(ints_arr)
                    if min(ints_arr) > 0.06: 
                        subdirs.append(subdir)
                        label_ =  str(label_)
                        for col in ["train_bd","val_bd","test_bd","train_llh","val_llh","test_llh","num_nodes"]:
                            label_ = label_ +"max_" + col + " = " + str(max(df_temp[col])) + " "
                        
                        ax.plot(df_temp["iteration"],df_temp[to_plot], label = label_)

        ax.legend(loc='lower right')

        plt.savefig(val+'.svg')

        return subdirs
```

```python
vals = ["train_bd","val_bd","test_bd","train_llh","val_llh","test_llh","num_nodes"]
for val in vals:
    plot_value(val,subdirs_)
```
